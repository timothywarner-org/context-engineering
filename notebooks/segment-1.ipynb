## Setup: anchor to the repo root

This cell is the reliability backbone. It walks **up** from the current working directory until it finds the `.git` marker, so every path below resolves correctly whether you launched the kernel at the repo root or inside `notebooks/`. One source of truth.

````python
# --- repo-root anchor (identical in all four segment notebooks) ---
import subprocess
from pathlib import Path


def find_repo_root(start: Path) -> Path:
    """Walk up from `start` until a directory containing .git is found."""
    for d in [start, *start.parents]:
        if (d / ".git").exists():
            return d
    raise RuntimeError("Repo root (.git) not found above " + str(start))


REPO = find_repo_root(Path.cwd())
LAB01 = REPO / "labs" / "lab-01-hello-mcp"
LAB02 = REPO / "labs" / "lab-02-mcp-chat"
WARNERCO = REPO / "src" / "warnerco" / "backend"

print("REPO     :", REPO)
print("LAB01    :", LAB01, "exists:", LAB01.exists())
assert LAB01.exists(), "Lab 01 not found - is REPO correct?"
````

## The `add` tool: name, registration, and TRUE return shape

The whole server is one tool. This cell extracts the actual registration from `solution/src/index.js` so you can see exactly what one MCP tool looks like - the name, and the **human-readable** return string (not `JSON.stringify`).

````python
# Extract the real 'add' tool from the JS source (verifiable, no guessing)
import re

src = (LAB01 / "solution" / "src" / "index.js").read_text(encoding="utf-8")

name = re.search(r"name:\s*'([^']+)'", src)
print("Server name   :", name.group(1) if name else "?")

tool = re.search(r"server\.tool\(\s*'([^']+)'", src)
print("Registered tool:", tool.group(1) if tool else "?")

text_line = re.search(r"text:\s*(`[^`]+`)", src)
print("\nReturned text  :", text_line.group(1) if text_line else "(not found)")
print("\nNote: returns a template string, NOT JSON.stringify(result).")
````
